---

```python
# Filtrar DataFrames
# El filtrado se aplica a las filas
filtro = datos.UsuariosInternet < 2
filtro
```

```python
datos[filtro]
```

```python
# Para filtrar datos se puede crear una variable
filtro2 = datos.TasaNatalidad > 40
datos[filtro2]
```

```python
# Pero también sin necesidad de crear una variable
datos[datos.TasaNatalidad > 40]
```

```python
# Combinar los dos filtros
datos[filtro and filtro2]   # Error al esperar un solo True o False de cada variable
```

```python
datos[filtro & filtro2]   # Con este operador analiza valor por valor
```

```python
datos[(datos.TasaNatalidad > 40) & (datos.UsuariosInternet < 2)]    # Entre paréntesis para expresar prioridades
```

```python
# Extracción de datos según valor de columna
datos[datos.GrupoIngresos == "Ingreso alto"]
```

```python
# Para comprobar todos los valores que hay en una columna
datos.GrupoIngresos.unique()
```

```python
# Encontrar todos los datos acerca de un país
datos[datos.NombrePais == "Malta"]
```

```python
# Utilizaremos las siguientes funciones
# .at     para etiquetas
# .iat    para ubicación con números enteros
```

```python
# Otro ejemplo con diferencias entre las dos funciones
sub10 = datos[ : : 10]
sub10
```

```python
sub10.iat[10, 0]    # .iat cuenta 10 posiciones
```

```python
sub10.at[10, "NombrePais"]   # .at considera el número 10 una etiqueta
```
